```python
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.orm import declarative_base, sessionmaker

# 1. Create database connection
engine = create_engine('sqlite:///data1.db', echo=True)

# 2. Create Base class
Base = declarative_base()

# 3. Define table (Model)
class User(Base):
    __tablename__ = 'users'   # âœ… FIXED

    id = Column(Integer, primary_key=True)
    name = Column(String)
    email = Column(String)
    course = Column(String)

# 4. Create table
Base.metadata.create_all(engine)

# 5. Create session
Session = sessionmaker(bind=engine)
session = Session()

# 6. Insert data
user1 = User(name="Laxmi", email="laxmi@gmail.com", course="Python")
user2 = User(name="Ram", email="ram@gmail.com", course="SQL")

session.add_all([user1, user2])
session.commit()

# 7. Read data
users = session.query(User).all()

for user in users:
    print(user.id, user.name, user.email, user.course)
```
