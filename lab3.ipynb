#Creating Base class Employee and defining constructor with private variables and Getting methods (encapsulation) and setting methods for salary

```python
class Employee:
    def __init__(self, employee_code, full_name, monthly_pay):
        self.__employee_code = employee_code
        self.__full_name = full_name
        self.__monthly_pay = monthly_pay
        
    # Getting Methods    
    def get_employee_code(self):
        return self.__employee_code
    
    def get_full_name(self):
        return self.__full_name
    
    def get_monthly_pay(self):
        return self.__monthly_pay
    
    # Setting Method
    def set_monthly_pay(self, monthly_pay):
        self.__monthly_pay = monthly_pay
```


#Defining Child class Professor that inherites the property of Base class Employee and accessing the private variables through super() function in Child class

```python
class Professor(Employee):
    def __init__(self, employee_code, full_name, monthly_pay, teaching_area):
        super().__init__(employee_code, full_name, monthly_pay)
        
        self.teaching_area = teaching_area
        
    # Display function
    def display_details(self):
        print("Professor Details")
        print("------------------")
        print("ID:", self.get_employee_code())
        print("Name:", self.get_full_name())
        print("Salary:", self.get_monthly_pay())
        print("Subject:", self.teaching_area)
```

Creating Object of professor(child) class

```python
professor_obj = Professor(1, "laxmi", 141000, "OOP")

# Calling display function with object
professor_obj.display_details()
```

File handling with txt

```python
student_records = [
    "1, Subekshya, 85",
    "2, Sube, 90",
    "3, Laxmi, 78"
]

with open("students.txt", 'w') as text_file:
    for entry in student_records:
        text_file.write(entry)
```

Read data from txt file

```python
with open("students.txt", "r") as text_file:
    file_content = text_file.read()

print("\nReading data from file:")
print(file_content)
```
