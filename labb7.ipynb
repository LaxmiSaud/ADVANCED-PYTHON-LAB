```python
import numpy as np
import pandas as pd 

data ={
  'Math': [85, 90, 78, 92],
    'Science': [88, 94, 80, 91],
 }
df = pd.DataFrame(data)
print(df)
```

```python
cov_matrix = df.cov()
print(cov_matrix)   
```

```python
corr_matrix = df.corr(method='pearson')
print(corr_matrix)
```

```python
import pandas as pd
import statsmodels.api as sm

data ={
    'hours_studied': [85, 90, 78, 92],
        'Score': [88, 94, 80, 91],
     }
df = pd.DataFrame(data)

X = df['hours_studied']
y = df['Score']

X = sm.add_constant(X)  # Adds a constant term to the predictor
model = sm.OLS(y, X).fit()
print(model.summary())
```
