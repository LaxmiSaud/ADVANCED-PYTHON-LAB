Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is the process of examining, summarizing, and visualizing a dataset to understand its main characteristics before applying statistical models or machine learning algorithms.

It helps to:

Understand the structure of data.Detect missing values and outliers.Identify patterns and relationships between variables


* Objectives of EDA

To gain insights into the dataset.

To clean and preprocess data.

To detect anomalies or outliers.

To select important features for modeling.

To improve decision-making and model accuracy.

THEORY

R-squared (R²):
R-squared measures how well a regression model explains the variability of the dependent variable using the independent variables. It represents the proportion of variance in the outcome that can be predicted from the inputs, with values ranging from 0 to 1. A higher R-squared indicates that the model fits the data better, meaning more of the variation is explained by the predictors, while a lower value suggests the model does not capture the relationship well.

Coefficients:
Coefficients are the numerical values assigned to each independent variable in a regression equation, showing how much the dependent variable is expected to change when the independent variable increases by one unit, assuming other variables remain constant. They also indicate the direction of the relationship: a positive coefficient means the dependent variable increases as the predictor increases, while a negative coefficient means it decreases.

P-value:
The p-value is used to determine the statistical significance of each independent variable in the model. It helps assess whether the observed relationship between a predictor and the dependent variable is likely real or occurred by chance. A commonly used threshold is 0.05; if the p-value is less than 0.05, the variable is considered statistically significant, meaning it has a meaningful contribution to the model.

F-statistic:
The F-statistic evaluates the overall significance of the regression model by testing whether at least one independent variable has a relationship with the dependent variable. It compares the model with predictors to a model without predictors (intercept-only model). A large F-statistic along with a small associated p-value indicates that the model provides a better fit than a model with no predictors, meaning the regression is useful.

```python
import pandas as pd
import statsmodels.api as sm

data={
    'hours_studied':[1,2,3,4,5],
    'exam_score':[50,60,70,80,90]
}
df=pd.DataFrame(data)

X=df['hours_studied']
y=df['exam_score']

X=sm.add_constant(X)
model=sm.OLS(y,X).fit()
print(model.summary())
```
