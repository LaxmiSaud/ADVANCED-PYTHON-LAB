Defination and Syntax
Iterator:
An iterator is an object that allows you to traverse through all the elements of a collection (like lists, tuples, dictionaries, sets, etc.) one element at a time without needing to know the underlying structure.

It follows the iterator protocol, which requires two methods:

. __iter__() → Returns the iterator object itself.
. __next__() → Returns the next element from the sequence. Raises StopIteration when no more elements are available.

```python
#Syntax of iterator
class CountDown:
    def __init__(self, start):
        self.current = start

    def __iter__(self):
        return self  # The iterator object itself

    def __next__(self):
        if self.current <= 0:
            raise StopIteration
        num = self.current
        self.current -= 1
        return num

# Using the custom iterator
for num in CountDown(5):
    print(num)
```

Generator
Generators in Python are a special type of iterable that allow you to iterate over data lazily, meaning they generate items only when needed. This is particularly useful for handling large datasets or streams of data without consuming large amounts of memory.

```python
#Syntax
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a  # Yield the next value of the sequence
        a, b = b, a + b

# Usage example
for num in fibonacci(10):
    print(num)
```

Decorator
 Decorators are a powerful feature that allow you to modify or enhance the behavior of functions or classes without changing their actual code.
They are often used for logging, authentication, caching, performance measurement, and more.


```python
def decorator_function(original_function):
    def wrapper_function(*args, **kwargs):
        # Code to run before the original function
        print("Before function execution")
        
        result = original_function(*args, **kwargs)
        
        # Code to run after the original function
        print("After function execution")
        
        return result
    return wrapper_function
#decorator_function
def say_hello(name):
    print(f"Hello, {name}!")

say_hello("Alice")
```

An iterator is an object that remembers its current position while going through a collection (like a list, tuple, string, or dictionary).
It has two main methods:

.__iter__() → returns the iterator object itself

.__next__() → returns the next item in the sequence



```python
#Example of an iterator
my_list = [1, 2, 3]
it = iter(my_list)  # create an iterator

print(next(it))  # 1
print(next(it))  # 2
print(next(it))  # 3
# print(next(it))  # would raise StopIteration
```
